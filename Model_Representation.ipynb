# Problem Statement

As in the lecture, you will use the motivating example of housing price prediction.  
This lab will use a simple data set with only two data points - a house with 1000 square feet(sqft) sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000. These two points will constitute our *data or training set*. In this lab, the units of size are 1000 sqft and the units of price are 1000s of dollars.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

You would like to fit a linear regression model (shown above as the blue straight line) through these two points, so you can then predict price for other houses - say, a house with 1200 sqft.

```python
import numpy as np
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")

# using len() to find m
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
# access elements
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}) , y^({i})) = ({x_i} , {y_i})")
```

```python
# plot the data points

plt.scatter(x_train, y_train, marker='x', c ='r') # can also use color = r, c = red, color = red for color
# set the title
plt.title("Housing Prices")
# set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

for $x^{(0)}$, `f_wb = w * x[0] + b`

for $x^{(1)}$, `f_wb = w * x[1] + b`

```python
def generate_function_output(x, w, b):
  '''
  Generate predictions
  Inputs: m, b, x_train
  Output: array of predictions for each i
  '''
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w*x[i] + b

  return f_wb
```

```python
predictions = generate_function_output(x_train, w, b)
print(f"predictions = {predictions}")
```

```python
plt.plot(x_train, predictions, c = (1,0,1), label = 'Model Prediction')
plt.scatter(x_train, y_train, marker='x', c = 'r', label = 'Actual Values')

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Testing some of my ideas

test_range = np.arange(0, 250, 50)

prediction_list = [
    generate_function_output(x_train, w, b)
    for w in test_range
    for b in test_range
]

num = len(prediction_list)
for i, prediction in enumerate(prediction_list):
    plt.plot(x_train, prediction, c=(5/num, i/num, i/num), label='Model Prediction')

plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
```
